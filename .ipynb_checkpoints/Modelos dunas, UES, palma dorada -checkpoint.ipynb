```python
#separamos los datos de los 2 nodos
data_pd = data.query("devEui == @query_id")
data_dunas = data.query("devEui == @query_id2")
print(len(data_pd))
print(len(data_dunas))
```

```python
# fultramos los datos del periodo que queremos
fecha_inicio = '2022-09-21 00:00:00' # septiembre 21 2022
fecha_final = '2023-03-21 23:59:59' # marzo 21 2023

data_pd = data_pd.query("customer_tz_createdDateTime >= @fecha_inicio and customer_tz_createdDateTime <= @fecha_final")
data_dunas = data_dunas.query("customer_tz_createdDateTime >= @fecha_inicio and customer_tz_createdDateTime <= @fecha_final")
print(len(data_pd))
print(len(data_dunas))
```

# Removemos outliers de cada sensor
usamos el mÃ©todo IQR para remover outliers

## primero hacemos el proceso para **palma dorada**

```python
# primero visualizamos los datos
sns.boxplot(x=data_pd['pm2_5'])
```

```python
# definimos los quatiles y el iqr para palma dorada
q1 = data_pd['pm2_5'].quantile(0.25)
q3 = data_pd['pm2_5'].quantile(0.75)
iqr = q3- q1
q1, q3, iqr
```

```python
# definimos los limites para palma dorada
upper_limit = q3 + (1.5 * iqr)
lower_limit = q1 - (1.5 * iqr)
lower_limit, upper_limit
```

```python
# encontramos los outliers
data_pd.loc[(data_pd['pm2_5'] >= upper_limit) | (data_pd['pm2_5'] <= lower_limit)]
```

```python
#capping the data ceiling
new_data_pd = data_pd.copy()
new_data_pd.loc[(data_pd['pm2_5'] >= upper_limit), 'pm2_5'] = upper_limit
new_data_pd.loc[(data_pd['pm2_5'] <= lower_limit), 'pm2_5'] = lower_limit
print('data antes: ', len(data_pd))
print('data nueva: ', len(new_data_pd))
print('diferencia : ', len(data_pd) -len(new_data_pd))
new_data_pd
#imprimimos para serciosarnos de no perder valores
```

```python
sns.boxplot(x=new_data_pd['pm2_5'])
```

## ahora hacemos el proceso para **Dunas**

```python
# primero visualizamos los datos
sns.boxplot(x=data_dunas['pm2_5'])
```

```python
# definimos los quatiles y el iqr para dunas
q1 = data_dunas['pm2_5'].quantile(0.25)
q3 = data_dunas['pm2_5'].quantile(0.75)
iqr = q3 - q1
q1, q3, iqr
```

```python
# definimos los limites para dunas
upper_limit = q3 + (1.5 * iqr)
lower_limit = q1 - (1.5 * iqr)
lower_limit, upper_limit
```

```python
# encontramos los outliers
data_dunas.loc[(data_dunas['pm2_5'] >= upper_limit) | (data_dunas['pm2_5'] <= lower_limit)]
```

```python
#capping the data ceiling
new_data_dunas = data_dunas.copy()
new_data_dunas.loc[(data_dunas['pm2_5'] >= upper_limit), 'pm2_5'] = upper_limit
new_data_dunas.loc[(data_dunas['pm2_5'] <= lower_limit), 'pm2_5'] = lower_limit
print('data antes: ', len(data_dunas))
print('data nueva: ', len(new_data_dunas))
print('diferencia : ', len(data_dunas) -len(new_data_dunas))
new_data_dunas
#imprimimos para serciosarnos de no perder valores
```
